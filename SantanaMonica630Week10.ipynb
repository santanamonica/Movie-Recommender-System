```python
# Import package
import pandas as pd

# Importing the data for the movie ratings with the id's and timestamps
ratingsdata = pd.read_csv("Downloads/ratings.csv")
ratingsdata
```

```python
# Importing the data for the movie names / id's / genres

moviesdata = pd.read_csv("Downloads/movies.csv")
moviesdata
```

```python
# Merge the two datasets using pandas merge function and merge them on column "movieid"

ratingsdata = ratingsdata.merge(moviesdata, on = 'movieId', how = 'left')
ratingsdata.head(10)
```

```python
# Calculate the average ratings for all the movies using groupby() function and the mean() function

averageratings = pd.DataFrame(ratingsdata.groupby('title')['rating'].mean())
averageratings.head(10)
```

```python
# Calculate the total amount of ratings per movie using groupby() and count() functions

averageratings['TotalRatings'] = pd.DataFrame(ratingsdata.groupby('title')['rating'].count())
averageratings.head(10)
```

```python
# Pivot the table to have the userid as the rows, and the movie titles as the columns to start the modeling process

moviesinput = ratingsdata.pivot_table(index = 'userId', columns = 'title', values = 'rating')
moviesinput.head()
```

```python
# Remove warnings
import warnings
warnings.filterwarnings("ignore")

# Use the movierecs variable and the corrwith() function to correlate the movie of choice with every movie in the dataset

correlations = moviesinput.corrwith(moviesinput["(500) Days of Summer (2009)"])
```

```python
# Use the dropna() function to drop all NaN's and join the correlation and total ratings columns

movierec = pd.DataFrame(correlations, columns=['Correlation'])
movierec.dropna(inplace=True)
movierec = movierec.join(averageratings['TotalRatings'])
movierec.head()
```

```python
# Sort movies correlated with the user input by the movies with a total ratings over 100 to refine the model

over100ratings = movierec[movierec['TotalRatings'] > 100].sort_values('Correlation', ascending = False).reset_index()
over100ratings.head(10)
```

```python
# Merge the over 100 ratings dataset with the movies dataset on the column "title" so the "movieid" and "genres" is included

over100ratings = over100ratings.merge(moviesdata, on = 'title', how = 'left')
over100ratings.head(10)
```
